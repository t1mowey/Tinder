```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
# options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 15)

BASE_URL = "https://www.farfetch.com/sets/new-in-this-week-eu-men.aspx"
product_urls = set()

def collect_product_links():
    scroll_step = 300
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        for offset in range(0, last_height, scroll_step):
            driver.execute_script(f"window.scrollTo(0, {offset});")
            time.sleep(0.2)

        cards = driver.find_elements(By.CSS_SELECTOR, '[data-testid="productCard"] a[data-component="ProductCardLink"]')
        for card in cards:
            href = card.get_attribute("href")
            if href:
                product_urls.add(href)
                print('added href')

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

def go_to_next_page():
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, 'a[data-component="PaginationNextActionButton"]')
        if next_button.get_attribute("aria-disabled") == "false":
            next_url = next_button.get_attribute("href")
            if next_url.startswith("/"):
                next_url = "https://www.farfetch.com" + next_url
            return next_url
    except:
        pass
    return None


driver.get(BASE_URL + '?lang=en-US&country=US')
time.sleep(2)


while True:
    collect_product_links()
    print(f"Собрано {len(product_urls)} товаров...")

    next_page = go_to_next_page()
    if not next_page:
        break
    driver.get(next_page)
    time.sleep(2)

# Теперь проходим по каждой ссылке и открываем "The Details"
print(f"Всего уникальных товаров: {len(product_urls)}")

```

```python
from seleniumwire import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
import gzip

# Настройки драйвера
options = webdriver.ChromeOptions()
# options.add_argument("--headless")
options.add_argument("--window-size=1920,1080")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 15)

# URL товара
url = "https://www.farfetch.com/shopping/men/stone-island-padded-coat-item-28810135.aspx?storeid=10224&lang=en-US&country=US"
driver.get(url)
driver.execute_script("document.body.style.zoom='50%'")

# Кликаем "Details", чтобы инициировать нужный XHR
try:
    details_btn = wait.until(EC.presence_of_element_located(
        (By.XPATH, "//button[@data-component='AccordionButton'][.//p[contains(text(), 'Details')]]")
    ))

    if details_btn.get_attribute("data-expanded") != "true":
        driver.execute_script("arguments[0].scrollIntoView();", details_btn)
        time.sleep(0.5)
        details_btn.click()
        time.sleep(2)

except Exception as e:
    print("❌ Кнопка 'Details' не найдена или не нажата:", e)

# Перехват нужного запроса
target_url = "https://www.farfetch.com/experience-gateway"
matched_request = None

for request in driver.requests:
    if request.method == "POST" and target_url in request.url:
        try:
            body = request.body.decode()
            if '"operationName":"SizeAndFitData"' in body:
                matched_request = request
                print("✅ Найден запрос SizeAndFitData!")
                break
        except Exception as e:
            print("⚠️ Ошибка при разборе тела запроса:", e)
driver.quit()

# Выводим информацию о запросе
if matched_request:
    print("🔗 URL:", matched_request.url)
    print("\n📤 Request Headers:")
    print(json.dumps(dict(matched_request.headers), indent=2))

    print("\n📦 Payload:")
    try:
        print(json.dumps(json.loads(matched_request.body.decode()), indent=2))
    except:
        print(matched_request.body.decode())

    if matched_request.response:
        print("\n📥 Response Status:", matched_request.response.status_code)
        print("\n📥 Response Headers:")
        print(json.dumps(dict(matched_request.response.headers), indent=2))

        try:
            print("\n📥 Response Body (decompressed):")
            try:
                raw = matched_request.response.body
                decompressed = gzip.decompress(raw).decode("utf-8")
                print(json.dumps(json.loads(decompressed), indent=2))
            except Exception as e:
                print("❌ Не удалось распаковать ответ:", e)
        except:
            print(matched_request.response.body.decode())
    else:
        print("⚠️ Нет ответа (response is None)")
else:
    print("❌ Не найден запрос к experience-gateway")
```

```python
from seleniumwire import webdriver
import requests

# --- Сначала заходишь через Selenium ---
options = webdriver.ChromeOptions()
# options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

driver.get("https://www.farfetch.com/shopping/men/prada-wool-bomber-jacket-item-29768955.aspx?storeid=11241")

# Ждём немного, чтобы всё успело загрузиться
driver.implicitly_wait(5)

# --- Получаем куки и заголовки ---
selenium_cookies = {c['name']: c['value'] for c in driver.get_cookies()}
headers = {
    "User-Agent": driver.execute_script("return navigator.userAgent;"),
    "Accept": "*/*",
    "Referer": "https://www.farfetch.com/",
    "Accept-Language": "en-US,en;q=0.9",
}

driver.quit()

# --- Эмулируем запрос через requests ---
session = requests.Session()
for k, v in selenium_cookies.items():
    session.cookies.set(k, v)

response = session.get(
    "https://www.farfetch.com/shopping/men/prada-wool-bomber-jacket-item-29768955.aspx?storeid=11241",
    headers=headers
)

print(response.status_code)
print(response.text[:1000])  # покажет HTML
```
